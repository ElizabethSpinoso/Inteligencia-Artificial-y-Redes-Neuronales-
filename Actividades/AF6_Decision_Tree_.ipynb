1. Importar librerías

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer #buscar cómo cargar los datos de los que nosotros vamos a entrenar que en nuestro caso es breast cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import numpy as np
```

2. Carga de datos

```python
breast_cancer = load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
```

```python
len(y)
```

```python
y[0]
```

```python
breast_cancer.target_names
```

3. Preprocesamiento

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

4. Selección y entrenamiento del modelo

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
#Exactitud del modelo en el entrenamiento
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

5. Prueba del modelo

```python
y_pred = clf.predict(X_test)
```

```python
#Obtener la matriz de confusión
conf_matrix=confusion_matrix(y_test, y_pred)
```

```python
#Mostrar la matriz de confusión de resultados de la prueba
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['malignant', 'benign'], yticklabels=['malignant', 'benign'])
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.title('Matriz de Confusión (Prueba)')
plt.savefig("confusion_matrix.png", dpi=300)
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=breast_cancer.target_names))
```
